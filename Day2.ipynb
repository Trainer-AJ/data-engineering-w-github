```
%sql
-- how many left ?
select count(*) as total_emp from attempt_3.default.emp_attrition
```

```
%sql
select sum(employeecount) as num_of_emp, attrition as Left_Job 
from attempt_3.default.emp_attrition
group by Attrition
```

```
%sql
-- Age Analysis - Which age group attrition is high
select min(age) as min_age, max(age) as max_age
from attempt_3.default.emp_attrition
```

```
%sql
select sum(employeecount) as num_of_emp, 
case when age between 18 and 25 then 'Below 20' when age between 26 and 32 then '26-32' when age between 33 and 40 then '33-40' else 'more than 40' end age_group
from attempt_3.default.emp_attrition
where Attrition = 'Yes'
group by 2 order by num_of_emp DESC
```

```
%sql
-- attrition by department
select sum(employeecount) as employee_who_left, Department
from attempt_3.default.emp_attrition
where Attrition = 'Yes'
group by 2 order by 1 DESC
```
